# Import Library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Loading & Preprocessing

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/DR21-ds/Breast-Cancer-Prediction/main/breast-cancer.csv')
```

```python
pd.set_option('display.max_columns',None)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

## Label Encoding 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lc = LabelEncoder()
df['diagnosis'] = lc.fit_transform(df['diagnosis'])
```

```python
df.head()
```

## Selecting Feature 

```python
temp =df.corr()
temp
```

```python
hig_corr_features = temp.index[abs(temp["diagnosis"]) >= 0.5]
hig_corr_features
```

```python
low_corr_features = set(temp.index)-set(hig_corr_features)
low_corr_features.add('diagnosis')
low_corr_features
```

```python
plt.figure(figsize=(10,8))
ax = sns.heatmap(df[hig_corr_features].corr(), annot=True, linewidth=3)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[low_corr_features].corr(),annot=True, linewidth=1.5)
```

- Correlation between radius_mean and [perimeter_mean,area_mean,radius_worst,perimeter_worst,area_worst] are almost One.
- Correlation between radius_se and [perimeter_se,area_se] are almost One.

```python
feature = ['radius_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
       'radius_se','compactness_worst', 'concavity_worst', 'concave points_worst']
plt.figure(figsize=(12,12))
for i in range(len(feature)):
        plt.subplot(3,3,i+1)
        plt.subplots_adjust(hspace = 0.5, wspace = 0.5)
        sns.regplot(data=df, x = feature[i], y = 'diagnosis')
```

```python
plt.figure(figsize=(12,12))
for i,j in enumerate(feature):
    plt.subplot(3,3,i+1)
    plt.subplots_adjust(hspace = 0.5, wspace = 0.5)
    ax = sns.distplot(df[j])
```

## Seleceted Feature Scaling

```python
df_n = df[feature]
df_n.shape
```

```python
from mlxtend.preprocessing import minmax_scaling
```

```python
df_n = minmax_scaling(df_n,columns=feature)
```

```python
plt.figure(figsize=(12,12))
for i,j in enumerate(feature):
    plt.subplot(3,3,i+1)
    plt.subplots_adjust(hspace = 0.5, wspace = 0.5)
    ax = sns.distplot(df_n[j])
```

# Traininig and Testing 

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_n
y = df['diagnosis']
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.1, stratify=y, random_state=1)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_train)
acc_training_data=accuracy_score(y_train, y_pred)
print('Accuracy on Training Data : ' ,round(acc_training_data*100,2))
```

```python
y_pred_t = classifier.predict(x_test)
acc_test_data=accuracy_score(y_test, y_pred_t)
print('Accuracy on Test Data : ' ,round(acc_test_data*100,2))
```

# K Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_classifier = KNeighborsClassifier()
k_classifier.fit(x_train,y_train)
```

```python
k_pred = k_classifier.predict(x_train)
k_acc_training_data=accuracy_score(y_train, k_pred)
print('Accuracy on Training Data : ' ,round(k_acc_training_data*100,2))
```

```python
k_pred_t= k_classifier.predict(x_test)
k_acc_testing_data=accuracy_score(y_test, k_pred_t)
print('Accuracy on Testing Data : ' ,round(k_acc_testing_data*100,2))
```

## Evaluation of Model with hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = {'n_neighbors' : [3,5,7,9,11,12,13,15,17,19],
          'weights' : ['uniform','distance'],
          'leaf_size' : [10,20,30,40,50],
          'p' : [1,2]}

random_cv = RandomizedSearchCV(estimator=k_classifier,
            param_distributions=params,
            cv=5, n_iter=50,
            n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
```

```python
random_cv.fit(x_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
k_classifier = KNeighborsClassifier(n_neighbors=17, weights='distance')
k_classifier.fit(x_train,y_train)

k_pred = k_classifier.predict(x_train)
k_acc_training_data=accuracy_score(y_train, k_pred)
print('Accuracy on Training Data : ' ,round(k_acc_training_data*100,2))

k_pred_t= k_classifier.predict(x_test)
k_acc_testing_data=accuracy_score(y_test, k_pred_t)
print('Accuracy on Testing Data : ' ,round(k_acc_testing_data*100,2))
```
